 ## Data Analysis and Visualization

```python
cancelled_perc = df['is_canceled'].value_counts(normalize=True)
cancelled_perc
```

#### this cancellation percetage (37%) is way much higher

```python
plt.figure(figsize=(5,4))
plt.title('Reservation Status Count')
plt.bar(['Not Canceled','Canceled'],df['is_canceled'].value_counts())
plt.show()
```

```python
#use countplot to find the cancellation rate according to the hotels

plt.figure(figsize=(8,4))
ax1=sns.countplot(x='hotel', hue='is_canceled',data=df,palette='Reds')
ax1.legend(title = 'Reservation Status',loc='upper right', labels=['Not Canceled','Canceled'])
plt.title('Reservation Status in different Hotels', size=20)
plt.xlabel('hotel')
plt.ylabel('number of reservations')
plt.show()
```

It is clearly visible that comparatively the cancelation in resort hotels is lesser than in city hotels.
What may be the reason? Price? Let's see....

```python
#calculate percentage of cancellations in resort and city hotels - 

res= df[df['hotel']=='Resort Hotel']
res['is_canceled'].value_counts(normalize=True)
```

```python
# for city hotels

city= df[df['hotel']=='City Hotel']
city['is_canceled'].value_counts(normalize=True)
```

```python
#to check the effect of price, take mean of average daily rate

res = res.groupby('reservation_status_date')[['adr']].mean()
city = city.groupby('reservation_status_date')[['adr']].mean()
```

```python
plt.figure(figsize=(20,8))
plt.title('Average Daily Rate in City and Resort Hotels', fontsize = 30)
plt.plot(res.index,res['adr'], label='Resort Hotel')
plt.plot(city.index,city['adr'], label='City Hotel')
plt.legend(fontsize=20)
plt.show()
```

spikes in the rate are at weekends when the price for both increases


```python
#month with max reservations and cancellations

df['month']=df['reservation_status_date'].dt.month
plt.figure(figsize=(16,8))
ax1=sns.countplot(x='month',hue = 'is_canceled',data=df, palette='bright')
ax1.legend(title = 'Reservation Status',loc='upper right', labels=['Not Canceled','Canceled'])
plt.title('Reservation Status per Month',size=20)
plt.xlabel('month')
plt.ylabel('number of reservations')
plt.legend(['not canceled', 'canceled'])
plt.show()
```

now this is quite confusing that why in the month when we have the most reservations, the cancellatio rate is least, and in the mosth of least reservation, cancellation rate is high? One reason may be that in august time the hotel price was high that's why people didn't cancel



```python
plt.figure(figsize=(15,8))
plt.title('ADR per month', fontsize=30)
sns.barplot('month','adr',data=df[df['is_canceled']==1].groupby('month')[['adr']].sum().reset_index())
plt.show()
```

so when higher price, cancellation will be higher because it is costly, and when price is least, the cancellation is least.

now let's see if the adr(price) is higher for cancelled reservations or not cancelled reservations?

```python
cancelled_df_adr = cancelled_data.groupby('reservation_status_date')[['adr']].mean()
cancelled_df_adr.reset_index(inplace=True)
cancelled_df_adr.sort_values('reservation_status_date',inplace=True)

not_cancelled_data = df[df['is_canceled']==0]
not_cancelled_df_adr = not_cancelled_data.groupby('reservation_status_date')[['adr']].mean()
not_cancelled_df_adr.reset_index(inplace=True)
not_cancelled_df_adr.sort_values('reservation_status_date',inplace=True)

plt.figure(figsize=(20,6))
plt.title('Average Daily Rate')
plt.plot(not_cancelled_df_adr['reservation_status_date'],not_cancelled_df_adr['adr'], label = 'not cancelled')
plt.plot(cancelled_df_adr['reservation_status_date'],cancelled_df_adr['adr'], label = 'cancelled')
plt.legend()
```

since there is inconsistencyy in data because we have lesser data before 2015 and after 2017, so we will now plot only for 2016

```python
cancelled_df_adr = cancelled_df_adr[(cancelled_df_adr['reservation_status_date']>'2016') & (cancelled_df_adr['reservation_status_date']<'2017-09')]
not_cancelled_df_adr = not_cancelled_df_adr[(not_cancelled_df_adr['reservation_status_date']>'2016') & (not_cancelled_df_adr['reservation_status_date']<'2017-09')]
```

```python
plt.figure(figsize=(20,6))
plt.title('Average Daily Rate', fontsize=20)
plt.plot(not_cancelled_df_adr['reservation_status_date'],not_cancelled_df_adr['adr'], label = 'not cancelled')
plt.plot(cancelled_df_adr['reservation_status_date'],cancelled_df_adr['adr'], label = 'cancelled')
plt.legend()
```

so here, average daily rate of cancelled is higher than that of not cancelled, hence price plays major role in the cancellation and non-cancellation

spikes are just seen on the weekends or month ends or holidays
